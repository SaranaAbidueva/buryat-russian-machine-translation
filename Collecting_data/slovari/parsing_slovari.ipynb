```python
list_ = str_.split('\n')
```

```python
sokr_list = []
for str_ in list_:
    word = str_.split('-')[0]
    word = word[:-1]
    sokr_list.append(' ' + word)
sokr_list.append('побудит,')
```

```python
good_words = []
all_words = []
not_matched = []
bad_words_v1 = []
bad_words_v2 = []
for page in text:
    page = re.split(r'[\.\?\!]\n', page)
    for word in page:
        # убираем сокращения
        if re.findall(r'частица', word):
            continue
        elif re.findall(r'межд.', word):
            continue
        for sokr in sokr_list:
            word = word.replace(sokr, '')
        # убираем номер страницы и первые буквы в начале страницы
        word = re.sub(r'[А-Я]+ +\-[\d+ +]+\- [А-Я]+(\n)+', '', word)
        # убираем все, что в скобках
        word = re.sub(r'[\(\{][^()]*[\)\}]', '', word)
        # убираем спецсимволы
        word = word.replace('\xad\n', '')
        word = word.replace('\n', '')
        word = word.split(';')[0]
        # убираем от какого слово произошло слово
        word = re.sub(r' от +[а-яёһһүөҺҺ]+', '', word)
        all_words.append(word)
    
        if re.fullmatch(r'([а-яёһһүөҺҺ]+ +[а-яё]+){1} *', word):
            good_words.append(word)
        elif re.match(r'([а-яёһһүөҺҺ]+ +[а-яё]+\, +[а-яё]+){1}', word):
            good_words.append(word)
        # 1)
        elif re.match(r'([а-яёһһүөҺҺ]+ +1\) +[а-яё]+)', word):
            good_words.append(word.replace('1)', ''))
        # 1.
        elif re.match(r'([а-яёһһүөҺҺ]+ +1\. +[а-яё]+)', word):
            good_words.append(word.replace('1.', ''))
        # I
        elif re.match(r'([а-яёһһүөҺҺ]+ +I +[а-яё]+)', word):
            good_words.append(word.replace('I', ''))
        # I)
        elif re.match(r'([а-яёһһүөҺҺ]+ +I\) +[а-яё]+)', word):
            good_words.append(word.replace('I)', ''))
        # 1 )
        elif re.match(r'([а-яёһһүөҺҺ]+ +1 \) +[а-яё]+)', word):
            good_words.append(word.replace('1 )', ''))
        # I 1)
        elif re.match(r'([а-яёһһүөҺҺ]+ +I +1\) +[а-яё]+)', word):
            good_words.append(word.replace('I 1)', ''))
        # 1. 1) 
        elif re.match(r'([а-яёһһүөҺҺ]+ +1\. +1\) +[а-яё]+)', word):
            good_words.append(word.replace('1. 1) ', ''))
        elif re.match(r'([а-яёһһүөҺҺ]+ +I 1\. +[а-яё]+)', word):
            good_words.append(word.replace('I 1.', ''))
            
        
        elif re.match(r'([а-яёһһүөҺҺ]+ +II +[а-яё]+)', word):
            good_words.append(word.replace('II', ''))
        elif re.match(r'([а-яёһһүөҺҺ]+ +II +1\) +[а-яё]+)', word):
            good_words.append(word.replace('II', ''))
        elif re.match(r'([а-яёһһүөҺҺ]+ +II I\) +[а-яё]+)', word):
            good_words.append(word.replace('II I)', ''))
            
        elif re.match(r'(^[а-яёһһүөҺҺ]+\,)', word):
            bur_word = re.findall(r'(^[а-яёһһүөҺҺ]+\,)', word)[0][:-1]
            rus_word = word.split()[-1]
            bad_words_v1.append(bur_word + ' ' + rus_word)
            if len(word.split(',')[1].split()) > 1:
                bad_words_v2.append(word.split(',')[1])
            elif len(word.split(',')) > 2:
                if len(word.split(',')[2].split()) > 1:
                    bad_words_v2.append(word.split(',')[2])
                    
        elif re.fullmatch(r'([а-яёһһүөҺҺ]+ +[а-яё]+ +[а-яё]+){1} *', word):
            good_words.append(word)
        elif re.fullmatch(r'([а-яёһһүөҺҺ]+ +[а-яё]+ +[а-яё]+ +[а-яё]+){1} *', word):
            good_words.append(word)
        else:
            not_matched.append(word)
```
