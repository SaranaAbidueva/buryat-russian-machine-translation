```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'http://soyol.ru/substrings/4387/'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html.parser')
```

```python
items = soup.find_all(r'i')
rus = []
for i in items:
    strings = i.text.strip().split('\n')
    rus += strings
```

```python
all_text = soup.find('div', class_="text-padding")
wrong_items = all_text.find_all('small')
text = soup.find('div', class_="text-padding").text
```

```python
text_list = text.strip().split('\n')
```

```python
wrong_text = []
for i in wrong_items:
    wrong_text.append(i.text.strip())
```

```python
text_list_bur = [s for s in text_list if s not in wrong_text and s not in rus]
```

```python
len(set(text_list_bur))
```

```python
len(rus)
```

```python
rus = [el for el in rus if el!='\r' and el!='']
len(rus)
```

```python
lost_strings = '''На груди широкого неба<br>
Солнце золотое блистает.<br>
И связывая времена<br>
Дни и ночи считает.<br>
Солнца лучом, солнца лучом,<br>
Солнца лучом вскормлена Земля.<br>
Мирного времени, мирного времени<br>
Мирного времени богата колыбель.<br>
Серебряным инеем звезд<br>
Привет передает весне.<br>
Любви волной, любви волной,<br>
Любви волной приплывает весна.<br>
Чащей полной, чащей полной,<br>
Чащей полной подается счастье.<br>
Лазурит драгоценный судьбы<br>
По безбрежной вселенной плывет.<br>
Стрелой священной любви<br>
Только раз можно разить.<br>
Судьбы такой, судьбы такой,<br>
Судьбы такой счастье испытать.<br>
Отца, матери, отца, матери<br>
Отца матери благословенья принять.<br>'''
```

```python
ind = rus.index('Связь времен')
```

```python
lost_str_list = lost_strings.split('<br>\n')
```

```python
rus = rus[:ind] + lost_str_list + rus[ind+1:]
```

```python
with open('rus.txt', 'w', encoding='utf-8') as fp:
    for item in rus:
        fp.write("%s\n" % item)
    print('Done')
```

```python
ind1 = text_list_bur.index('На\xa0груди широкого неба\r')
ind2 = ind1 + len(lost_str_list) 
del text_list_bur[ind1:ind2]
```

```python
len(text_list_bur)
```

```python
reg = r'^\*'
bur = []
for s in text_list_bur:
    if re.search(reg, s) is None:
        bur.append(s)
```
