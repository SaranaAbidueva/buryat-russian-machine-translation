```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
def split_sentences(text):
    split_text = [] 
    string = ''
    for i in text:
        string += i + ' '
    reg = r'(?<=[\.\?\!])'
    # reg = r'(?<=[\.\?\!])(?!(»|“))'
    new_lines = re.split(reg, string)
    for line in new_lines:
        if line is not None:
            if re.findall(r'(\w|[«„\d])+', line)!=[] and re.findall(r'(\d\:\d)+', line)==[]:
                split_text.append(line.lstrip())
    return split_text
```

```python
def parse(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    try:
        items = soup.find_all(r'i')
        rus = []
        for i in items:
            strings = i.text.strip().split('\n')
            rus += strings
        all_text = soup.find('div', class_="text-padding")
        wrong_items = all_text.find_all('small')
        text = soup.find('div', class_="text-padding").text
        text_list = text.strip().split('\n')
        wrong_text = []
        for i in wrong_items:
            wrong_text.append(i.text.strip())
        text_list_bur = [s for s in text_list if s not in wrong_text and s not in rus]
        rus = [el for el in rus if el!='\r' and el!='']
        reg = r'^\*'
        bur = [rus[0]]
        for s in text_list_bur:
            if re.search(reg, s) is None:
                bur.append(s)
        bur = split_sentences(bur)
        rus = split_sentences(rus)
        print('bur ', len(bur))
        print('rus ', len(rus))
        print(url)
        for i in range(len(bur)):
            line = bur[i]
            bur[i] = line.replace('\r', '').replace('\xa0', ' ')
        for i in range(len(rus)):
            line = rus[i]
            rus[i] = line.replace('\r', '').replace('\xa0', ' ')
        return rus, bur
    except AttributeError:
        pass
   
```

```python
ids = range(4365, 4387)
for i in ids:
    url = f'http://soyol.ru/substrings/{i}/'
    parse(url)
```
