# библия на бурятском

```python
path_bur = 'bible/Шэнэ-Хэлсээн__Full.txt'
f = open(path_bur,'r')
```

```python
def read_text(path):
    text = []
    with open (path, 'rt', encoding="utf8") as file: 
        for line in file:                
            text.append(line)
    return text
raw_bur = read_text(path_bur)
```

```python
def remove_spaces(text):
    text_wo_spaces = []
    for line in text:
        if line!='\n':
            text_wo_spaces.append(line.strip())
    return text_wo_spaces
bur_wo_spaces = remove_spaces(raw_bur)
bur_wo_spaces[:10]
```

```python
#split on sentences
def split_sentences(text):
    split_text = [] 
    for i in range(len(text)):
        reg = r'(?<=[\.\?\!])'
        # reg = r'(?<=[\.\?\!])(?!(»|“))'
        new_lines = re.split(reg, text[i])
        for line in new_lines:
            if line is not None:
                if re.findall(r'(\w|[«„\d])+', line)!=[] and re.findall(r'(\d\:\d)+', line)==[]:
                    split_text.append(line.lstrip())
    return split_text
text = split_sentences(bur_wo_spaces)
```

```python
def remove_unwanted_symbols(text):
    for i in range(len(text)):
        line = text[i]
        if re.findall(r'(•|\+)', line):
            text[i] = re.sub(r'(•|\+)', '', line)
    return text
```

```python
text_wo_symbols = remove_unwanted_symbols(text)
text_wo_symbols[80:100]
```

```python
def split_digits(text):
    new_text = [' ']
    reg = r'\d+'
    j = 1  # index of new_text
    for i in range(len(text)):
        line = text[i]
        if re.findall(reg, line):
            new_text.append(line)
            j += 1
        elif new_text[j-1]:
            new_text[j-1] += (line + ' ')
        else:
            new_text[j-1] = line
    return new_text
```

```python
text = split_digits(text_wo_symbols)
text[:50]
```

```python
def split_more(text):
    reg = r'(\d+)'
    new_text = []
    all_numbers = []
    ranges = []
    for i in range(len(text)):
        line = text[i]
        found_numbers = re.findall(reg, line)
        all_numbers.append(found_numbers)
        # just 1 sentence
        if len(found_numbers) == 1:
            prev, number, line = line.partition(found_numbers[0])
            new_text.append(number + ' ' + line)
        # ranges
        elif len(found_numbers) > 1:
            if re.findall(fr'{found_numbers[0]}-{found_numbers[1]}+', line):
                if int(found_numbers[0]) > int(found_numbers[1]):
                    found_numbers[0] = found_numbers[0][-1]
                prev, number, line = line.partition(f'{found_numbers[0]}-{found_numbers[1]}')
                new_text.append(number + ' ' + line)
            else:
                s = ''
                for k in range(len(found_numbers)-1):
                    _, number, tail = line.partition(found_numbers[k])
                    prev, _,_  = tail.partition(found_numbers[k+1])
                    s += prev
                _, _, tail = line.partition(found_numbers[-1])
                s += tail
                new_text.append(f'{found_numbers[0]}-{found_numbers[-1]} {s}')                
            ranges.append([int(found_numbers[0]), int(found_numbers[-1])])
    return new_text
```

```python
new_text = split_more(text)
new_text[695:780]
```

```python
def remove_star_lines(text):
    new_text = []
    for i in range(len(text)):
        line = text[i]
        if re.findall(r'(\*)', line):
            continue
        else:
            new_text.append(line)
    return new_text
```

```python
def make_chapter_dict(text: list) -> dict:
    books = []
    book_number = 0    
    dictionary = {}
    lines = []
    prev_chapter = 2
    for i, line in enumerate(text):
        if re.findall(r'(=== \d+ ===)+', line)!=[]:
            found_number = int(re.findall(r'\d+', line)[0])
            if found_number < prev_chapter:
                book_number += 1
                books.append(book_number)
                dictionary[f'{book_number-1}.{prev_chapter}'] = lines
            else:
                dictionary[f'{book_number}.{prev_chapter}'] = lines
            lines = []
            prev_chapter = found_number
        else:
            lines.append(line)
    return dictionary
            
            
                
        
```

```python
def chapters(text):
    chapters = []
    arr = [0]
    names_chapters = []
    new_text = []
    for line in text:
        if re.findall(r'=+', line):
            chapter = re.findall(r'\d+', line)
            if chapter and int(chapter[0]) == 1:
                chapters.append(arr[-1])
                arr = [1]
            elif chapter:
                arr.append(int(chapter[0]))
            else:
                names_chapters.append(line)
                continue
        new_text.append(line)
    print(names_chapters)
    print(len(names_chapters))
    return new_text
```
