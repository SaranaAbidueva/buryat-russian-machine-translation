```python
with open(f'{base_path}parallel/slovar_bur_1.txt', 'r', encoding='utf-8') as f:
    bur_slovar = f.readlines()
with open(f'{base_path}parallel/slovar_rus_1.txt', 'r', encoding='utf-8') as f:
    rus_slovar = f.readlines()

with open(f'{base_path}parallel/slovar_bur_2_bad.txt', 'r', encoding='utf-8') as f:
    bur_slovar_2 = f.readlines()
with open(f'{base_path}parallel/slovar_rus_2_bad.txt', 'r', encoding='utf-8') as f:
    rus_slovar_2 = f.readlines()

with open(f'{base_path}parallel/bur_slovar_dop.txt', 'r', encoding='utf-8') as f:
    bur_slovar_dop = f.readlines()
with open(f'{base_path}parallel/rus_slovar_dop.txt', 'r', encoding='utf-8') as f:
    rus_slovar_dop = f.readlines()
rus_slova = rus_slovar + rus_slovar_2 + rus_slovar_dop
bur_slova = bur_slovar + bur_slovar_2 + bur_slovar_dop
# избавляюсь от дубликатов
list_tuple = []
for i in range(len(bur_slova)):
  list_tuple.append((bur_slova[i], rus_slova[i]))
slovar = list(set(list_tuple))
rus_slovar = []
bur_slovar = []
for i in range(len(slovar)):
  bur_slovar.append(slovar[i][0])
  rus_slovar.append(slovar[i][1])
len(rus_slovar), len(bur_slovar)
```

```python
bur_all = bur_bible + bur_books + bur_tatoeba + bur_stihi + bur_zakony + bur_wiki + bur_nimbuev
rus_all = rus_bible + rus_books + rus_tatoeba + rus_stihi + rus_zakony + rus_wiki + rus_nimbuev
len(rus_all), len(bur_all)
```

```python
bur_all += bur_slovar
rus_all += rus_slovar
```

```python
pairs = []
sources = ('библия', 'книги', 'татоеба', 'стихи', 'стихи Нимбуев', 'словарь', 'википедия', 'законы')
j = 0
for bur, rus in [(bur_bible, rus_bible), (bur_books, rus_books), (bur_tatoeba, rus_tatoeba), (bur_stihi, rus_stihi), (bur_nimbuev, rus_nimbuev), (bur_slovar, rus_slovar), (bur_wiki, rus_wiki), (bur_zakony, rus_zakony)]:
  pairs_temp = []
  for i in range(len(rus)):
    sent = tuple([bur[i].replace('\n', '').replace('һ', 'h'), rus[i].replace('\n', '').replace('\ufeff', '').replace('[', '').replace(']', ''), sources[j]])
    pairs_temp.append(sent)
  print(sources[j], ' ', len(set(pairs_temp)))
  pairs += set(pairs_temp)
  j += 1
print()
print('всего пар ', len(pairs))
```

# Собираю датасет в json, делю на train и test

```python
df = pd.DataFrame(list(pairs), columns= ['bxr', 'ru', 'source'])
```

```python
df.sample(5)
```

```python
train_df = pd.DataFrame(columns= ['bxr', 'ru', 'source'])
test_df = pd.DataFrame(columns= ['bxr', 'ru', 'source'])
sources = ('библия', 'книги', 'татоеба', 'стихи', 'стихи Нимбуев', 'википедия', 'законы')
for source in sources:
  train_i, test_i = train_test_split(df[df['source']==source], test_size=0.1, random_state=42)
  train_df = pd.concat([train_df, train_i], ignore_index=True)
  test_df = pd.concat([test_df, test_i], ignore_index=True)
```

```python
len(train_df), len(test_df)
```

```python
train_df = pd.concat([train_df, df[df['source']=='словарь']], ignore_index=True)
```

```python
len(train_df), len(test_df)
```

```python
len(train_df[train_df['source']=='библия'])
```

```python
bxr_train = list(train_df['bxr'])
ru_train = list(train_df['ru'])

bxr_test = list(test_df['bxr'])
ru_test = list(test_df['ru'])
```

```python
train = [(bxr_train[i], ru_train[i]) for i in range(len(bxr_train))]
test = [(bxr_test[i], ru_test[i]) for i in range(len(bxr_test))]
```

```python
with open("train.json", "w", encoding='utf-8') as outfile:
    json.dump(train, outfile)
```

```python
with open("test.json", "w", encoding='utf-8') as outfile:
    json.dump(test, outfile)
```
